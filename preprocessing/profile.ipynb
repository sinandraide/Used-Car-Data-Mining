```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import OrdinalEncoder

df_new = pd.read_csv("../data/train_data_preprocessed_no_outliers.csv")
numerical_features = df_new[["price", "year", "odometer", ]]

# Optional: If you have ordinal categorical features already encoded as numbers, include them
# For example:
encoder = OrdinalEncoder()

ordinal_features = encoder.fit_transform(df_new[['condition', 'size']])

# Convert to DataFrame with columns and index
ordinal_features = pd.DataFrame(ordinal_features, 
                                columns=['condition', 'size'], 
                                index=df_new.index)


numerical_features = pd.concat([numerical_features, ordinal_features], axis=1)
# Step 2: Calculate correlation matrix
corr_matrix = numerical_features.corr(method='pearson')  # or method='spearman'

# Step 3: Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```
